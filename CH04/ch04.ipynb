# <span class="header-section-number">1</span> Modules et _packages_

Un module est un fichier `.py`.

Définissons un module `calculs.py`:

    def calculer_taxes(montant, taux_tps=0.05, taux_tvq=0.09975):
        tps = montant * taux_tps
        tvq = montant * taux_tvq
        return tps + tvq

Utilisons le module à partir d’un autre module `pos.py`:

    import calculs

    print("Taxes =", calculs.calculer_taxes(9.95))

    $ python3 pos.py
    Taxes = 1.4900125

L’instruction `from ... import` permet d’importer directement des objets dans l’espace de noms (_namespace_):

    from calculs import calculer_taxes

    print("Taxes =", calculer_taxes(9.95))

Lorsqu’on importe un module, tout son code est exécuté par défaut. On peut exécuter un code de bloc seulement si le module est invoqué en tant que script:

    def main():
    ...

    if __name__ == '__main__':
        main()

Par exemple, lorqu’on exécute le module `http.server`, il exécute un serveur web. Voir: `/usr/lib/python3.4/http`.

## <span class="header-section-number">1.1</span> _Packages_

Un package est un répertoire regroupant plusieurs modules (fichiers `.py`) et le fichier `__init__.py`. Lors de l’importation d’un _package_, c’est ce fichier qui est exécuté.

Exemple d’un _package_ `report`:

    reports/
    |-- html.py
    |-- __init__.py
    |-- text.py
    |-- xml.py

Le fichier `__init__.py` pourrait ressembler à cela:

    import html
    import json
    import text
    import xml

    def print_report(report, format='text'):
        if format == 'html':
            html.print_report(report)
        elif format == 'text':
            text.print_report(report)
        elif format == 'xml':
            xml.print_report(report)

Un utilisateur du _package_ `reports` peut l’importer comme un module:

    from reports import print_report

    report = ...

    print_report(report, format='xml')
